```python
import os
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)
```

```python
rj = [x for x in os.listdir("data") if "rj_" in x]
df = pd.DataFrame()
for r in rj:
    df = pd.concat([df, pd.read_parquet(os.path.join("data",r))])
```
