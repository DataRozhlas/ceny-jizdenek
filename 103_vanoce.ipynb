```python
import os
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_parquet(os.path.join('data','jizdenky.parquet'))
```

```python
df = df.sort_values(by='oscrapovano')
df = df.drop_duplicates(subset='odjezd',keep='last')
```

```python
(df[df['volnych_mist'] <= 10].groupby(pd.Grouper(key='odjezd',freq='1D')).size() / df[df['volnych_mist'].notnull()].groupby(pd.Grouper(key='odjezd',freq='1D')).size()).plot.bar(figsize=(25,5))
```

```python
(df[df['volnych_mist'] <= 5].groupby(pd.Grouper(key='odjezd',freq='1W')).size() / df[df['volnych_mist'].notnull()].groupby(pd.Grouper(key='odjezd',freq='1W')).size()).plot.bar(figsize=(25,5))
```
