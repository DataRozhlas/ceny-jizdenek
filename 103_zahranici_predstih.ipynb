```python
import os
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_parquet(os.path.join('data','jizdenky.parquet'))
```

```python
df[(df['odkud'].str.contains('Prah')) & (df['kam'].str.contains('Bru')) & (df['prestupy'] == 0)]
```

```python
def predstih(odkud, kam, prodejce=['ČD','ARR','LE','RJ']):
    dfp = df[(df['prodejce'].isin(prodejce)) & (df['odkud'].str.contains(odkud)) & (df['kam'].str.contains(kam))]
    dfp['predstih_d2'] = dfp['predstih_d'].apply(lambda x: int(x))
    dfp = dfp.drop_duplicates(subset=["odjezd","predstih_d2"], keep="last")
    min_cas = dfp['jizdni_doba'].min()
    min_prestupu = dfp['prestupy'].min()
    print(dfp.groupby('cena')['predstih_d2'].min())
    dfp[(dfp['jizdni_doba'] <= min_cas * 1.1) & (dfp['prestupy'] == min_prestupu)].plot.scatter(x='predstih_d2',y='cena')
```

```python
predstih('Praha','Brati',prodejce=['ČD'])
```
